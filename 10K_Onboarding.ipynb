```python
def define_the_range(x, dic_list, bad_big = True):
    if not type(x) == int and not type(x) == float:
        return
    elif not bad_big:
        x = -x
        dic_list = [-a for a in reversed(dic_list)]
    if type(dic_list[0])==tuple:
        if x>=dic_list[0][0] and x<=dic_list[0][1]:
            return 'normal'
        elif x<dic_list[0][0] or x>dic_list[0][1]:
            return 'abnormal'

    elif len(dic_list)==1:
        if x<=dic_list[0]:
            return 'normal'
        elif x>dic_list[0]:
            return 'abnormal'
    elif len(dic_list)==2:
        if x<=dic_list[0]:
            return 'normal'
        elif x<=dic_list[1]:
            return 'intermediate'
        elif x>dic_list[1]:
            return 'abnormal'
```
